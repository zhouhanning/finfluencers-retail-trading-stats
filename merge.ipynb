```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("frankcaoyun/stocktwits-2020-2022-raw")

print("Path to dataset files:", path)
```

```python
import os
print(f"Path: {path}")
print("\nFiles in the directory:")
for item in os.listdir(path):
    print(f"  - {item}")
```

```python
import glob
# Find all CSV files
csv_files = glob.glob(os.path.join(path, "**/*.csv"), recursive=True)

# Filter AAPL and AMZN CSV files
aapl_files = [f for f in csv_files if 'AAPL' in os.path.basename(f)]
amzn_files = [f for f in csv_files if 'AMZN' in os.path.basename(f)]

print(f"Total CSV files: {len(csv_files)}")
print(f"AAPL CSV files: {len(aapl_files)}")
print(f"AMZN CSV files: {len(amzn_files)}")
print(f"\nFirst 5 AAPL files:")
for f in aapl_files[:5]:
    print(f"  - {os.path.basename(f)}")
print(f"\nFirst 5 AMZN files:")
for f in amzn_files[:5]:
    print(f"  - {os.path.basename(f)}")
```

```python
# Merge all AAPL CSV files
print("Reading and merging AAPL data...")
aapl_dfs = []
for i, file in enumerate(aapl_files):
    if i % 10 == 0:
        print(f"  AAPL Progress: {i}/{len(aapl_files)}")
    df = pd.read_csv(file)
    aapl_dfs.append(df)

aapl_stocktwits = pd.concat(aapl_dfs, ignore_index=True)
print(f"AAPL merge complete! Shape: {aapl_stocktwits.shape}")

# Merge all AMZN CSV files
print("\nReading and merging AMZN data...")
amzn_dfs = []
for i, file in enumerate(amzn_files):
    if i % 10 == 0:
        print(f"  AMZN Progress: {i}/{len(amzn_files)}")
    df = pd.read_csv(file)
    amzn_dfs.append(df)

amzn_stocktwits = pd.concat(amzn_dfs, ignore_index=True)
print(f"AMZN merge complete! Shape: {amzn_stocktwits.shape}")
```

```python
# View AAPL data
print("=" * 60)
print("AAPL StockTwits Data Overview")
print("=" * 60)
print(f"Shape: {aapl_stocktwits.shape}")
print(f"\nColumns:\n{list(aapl_stocktwits.columns)}")
print(f"\nFirst 5 rows:")
aapl_stocktwits.head()
```

```python
# First, let's examine the entities column structure
import json
print("Sample entities from AAPL:")
for i in range(5):
    if pd.notna(aapl_stocktwits.iloc[i]['entities']):
        print(f"Row {i}: {aapl_stocktwits.iloc[i]['entities']}")
        try:
            parsed = json.loads(aapl_stocktwits.iloc[i]['entities'].replace("'", '"'))
            print(f"  Parsed: {parsed}")
        except:
            print(f"  Could not parse")
        print()
```

```python
import ast

def extract_sentiment(entities_str):
    """Extract sentiment from entities column"""
    if pd.isna(entities_str):
        return None
    try:
        # Convert string to dict
        entities_dict = ast.literal_eval(str(entities_str))
        if 'sentiment' in entities_dict and entities_dict['sentiment'] is not None:
            if 'basic' in entities_dict['sentiment']:
                return entities_dict['sentiment']['basic']
    except:
        pass
    return None

# Apply sentiment extraction to both dataframes
print("Extracting sentiments from AAPL data...")
aapl_stocktwits['sentiment'] = aapl_stocktwits['entities'].apply(extract_sentiment)

print("Extracting sentiments from AMZN data...")
amzn_stocktwits['sentiment'] = amzn_stocktwits['entities'].apply(extract_sentiment)

print("\nSentiment distribution in AAPL:")
print(aapl_stocktwits['sentiment'].value_counts())
print("\nSentiment distribution in AMZN:")
print(amzn_stocktwits['sentiment'].value_counts())
```

```python
# Process AAPL data
# Convert created_at to datetime
aapl_stocktwits['created_at'] = pd.to_datetime(aapl_stocktwits['created_at'])

# Extract date from created_at
aapl_stocktwits['date'] = aapl_stocktwits['created_at'].dt.date

# Group by date and calculate sentiment counts
aapl_sentiment_daily = aapl_stocktwits.groupby('date')['sentiment'].value_counts().unstack(fill_value=0)

# Calculate net sentiment
if 'Bullish' in aapl_sentiment_daily.columns and 'Bearish' in aapl_sentiment_daily.columns:
    aapl_sentiment_daily['net_sentiment'] = (aapl_sentiment_daily['Bullish'] - aapl_sentiment_daily['Bearish']) / \
                                             (aapl_sentiment_daily['Bullish'] + aapl_sentiment_daily['Bearish'])
else:
    aapl_sentiment_daily['net_sentiment'] = 0

# Keep only net_sentiment column and reset index to make date a column
aapl_net_sentiment = aapl_sentiment_daily[['net_sentiment']].reset_index()
aapl_net_sentiment.columns = ['created_at', 'net_sentiment']

print("AAPL Net Sentiment:")
print(aapl_net_sentiment.head(10))
print(f"\nShape: {aapl_net_sentiment.shape}")
```

```python
# Process AMZN data
# Convert created_at to datetime
amzn_stocktwits['created_at'] = pd.to_datetime(amzn_stocktwits['created_at'])

# Extract date from created_at
amzn_stocktwits['date'] = amzn_stocktwits['created_at'].dt.date

# Group by date and calculate sentiment counts
amzn_sentiment_daily = amzn_stocktwits.groupby('date')['sentiment'].value_counts().unstack(fill_value=0)

# Calculate net sentiment
if 'Bullish' in amzn_sentiment_daily.columns and 'Bearish' in amzn_sentiment_daily.columns:
    amzn_sentiment_daily['net_sentiment'] = (amzn_sentiment_daily['Bullish'] - amzn_sentiment_daily['Bearish']) / \
                                             (amzn_sentiment_daily['Bullish'] + amzn_sentiment_daily['Bearish'])
else:
    amzn_sentiment_daily['net_sentiment'] = 0

# Keep only net_sentiment column and reset index to make date a column
amzn_net_sentiment = amzn_sentiment_daily[['net_sentiment']].reset_index()
amzn_net_sentiment.columns = ['created_at', 'net_sentiment']

print("AMZN Net Sentiment:")
print(amzn_net_sentiment.head(10))
print(f"\nShape: {amzn_net_sentiment.shape}")
```

```python
# Summary of processed data
print("=" * 70)
print("FINAL PROCESSED DATA SUMMARY")
print("=" * 70)

print("\nAAPL Net Sentiment DataFrame:")
print(f"  Shape: {aapl_net_sentiment.shape}")
print(f"  Columns: {list(aapl_net_sentiment.columns)}")
print(f"  Date Range: {aapl_net_sentiment['created_at'].min()} to {aapl_net_sentiment['created_at'].max()}")
print(f"  Net Sentiment Range: [{aapl_net_sentiment['net_sentiment'].min():.4f}, {aapl_net_sentiment['net_sentiment'].max():.4f}]")
print(f"  Mean Net Sentiment: {aapl_net_sentiment['net_sentiment'].mean():.4f}")

print("\nAMZN Net Sentiment DataFrame:")
print(f"  Shape: {amzn_net_sentiment.shape}")
print(f"  Columns: {list(amzn_net_sentiment.columns)}")
print(f"  Date Range: {amzn_net_sentiment['created_at'].min()} to {amzn_net_sentiment['created_at'].max()}")
print(f"  Net Sentiment Range: [{amzn_net_sentiment['net_sentiment'].min():.4f}, {amzn_net_sentiment['net_sentiment'].max():.4f}]")
print(f"  Mean Net Sentiment: {amzn_net_sentiment['net_sentiment'].mean():.4f}")

print("\n" + "=" * 70)
```
